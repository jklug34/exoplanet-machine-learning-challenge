```python
import pandas as pd
import numpy as np
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
# rename header to make easier to interpret what each column means
# see link below for detailed description for each header
# https://exoplanetarchive.ipac.caltech.edu/docs/API_kepcandidate_columns.html
df_cleaned_headers = pd.read_csv("exoplanet_data_headers.csv")
# Drop the null columns where all values are null
df_cleaned_headers = df_cleaned_headers.dropna(axis='columns', how='all')
# Drop the null rows
df_cleaned_headers = df_cleaned_headers.dropna()
df_cleaned_headers.head()
```

```python
columns = ['koi_period_err1', 'koi_period_err2', 'koi_time0bk_err1', 'koi_time0bk_err2', 'koi_impact_err1', 'koi_impact_err2', 'koi_duration_err1', 'koi_duration_err2', 'koi_depth_err1', 'koi_depth_err2', 'koi_prad_err1', 'koi_prad_err2', 'koi_insol_err1', 'koi_insol_err2', 'koi_steff_err1', 'koi_steff_err2', 'koi_slogg_err1', 'koi_slogg_err2', 'koi_srad_err1', 'koi_srad_err2', 'ra', 'dec', 'kepler_band_mag']
df_cleaned_headers.drop(columns, inplace=True, axis=1)
df_cleaned_headers.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df_cleaned_headers[['orbital_period_days', 'transit_epoch', 'impact_parameter', 'transit_duration_hours', 'transit_depth_ppm', 'planetary_radius', 'temperature_kelvin', 'insolation_flux', 'transit_signal_to_noise', 'tce_planet_number', 'stellar_effective_temperature_kelvin', 'stellar_surface_gravity', 'stellar_radius', 'not_transit_like_flag', 'stellar_eclipse_flag', 'centroid_offset_flag', 'ephemeris_match_contamination_flag']]
selected_features.head()
```

```python
selected_features.info()
```

```python
X = selected_features
y = df_cleaned_headers["koi_disposition"]
print(X.shape, y.shape)
```

```python
dataX = X.copy()
dataX_binary_encoded = pd.get_dummies(dataX)
dataX_binary_encoded.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split

print(X.shape, y.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)

X_train.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_minmax = MinMaxScaler().fit(X_train)


X_train_minmax = X_minmax.transform(X_train)
X_test_minmax = X_minmax.transform(X_test)
```

```python
# try logistic regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', max_iter=6000, multi_class='auto')
classifier
```

```python
# fit the confirmed y value
classifier.fit(X_train_minmax, y_train)
```

# Train the Model



```python
# train the model
print(f"Training Data Score: {classifier.score(X_train_minmax, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_minmax, y_test)}")
```

```python
# validate the stability of the model with cross-validation
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
# Build the k-fold cross-validator, 3-fold cross validator
kfold = KFold(n_splits=3, random_state=42)
```

```python
result = cross_val_score(classifier, X_test_minmax, y_test, cv=kfold, scoring='accuracy')
print(result.mean())
```

```python
predictions = classifier.predict(X_test_minmax)
```

```python
logistic_regression_model_pedict_df = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
logistic_regression_model_pedict_df.head()
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
from sklearn.model_selection import GridSearchCV
```

```python
max_iter=[500, 1000, 2000, 5000]
C= [300, 350, 400, 500, 600, 700]
#fit_intercept = [True, False]
#verbose = [0, 1, 2, 3]
#tol = [0.00001, 0.0001, 0.001]

param_grid = dict(max_iter=max_iter, C=C)
```

```python
import time

classifier = LogisticRegression(penalty='l2', solver='lbfgs', multi_class='auto')
grid = GridSearchCV(estimator=classifier, param_grid=param_grid, cv = 5, n_jobs=-1)

start_time = time.time()
grid_result = grid.fit(X_train_minmax, y_train)
# Summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
print("Execution time: " + str((time.time() - start_time)) + ' sec')
```

```python
classifier.get_params()
```

```python
# Make predictions with the hypertuned model
predictions = grid.predict(X_test_minmax)
predictions
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["CANDIDATE", "CONFIRMED", "FALSE POSITIVE"]))
```

```python
# try a random search
from sklearn.model_selection import RandomizedSearchCV
```

```python
max_iter=[500, 1000, 2000, 5000]
C= [300, 350, 400, 500, 600, 700]
#fit_intercept = [True, False]
#verbose = [0, 1, 2, 3]
tol = [0.00001, 0.0001, 0.001]

param_grid = dict(max_iter=max_iter, C=C, tol=tol)
```

```python
# same answer but runs faster
random = RandomizedSearchCV(estimator=classifier, param_distributions=param_grid, cv = 5, n_jobs=-1)

start_time = time.time()
random_result = random.fit(X_train_minmax, y_train)
# Summarize results
print("Best: %f using %s" % (random_result.best_score_, random_result.best_params_))
print("Execution time: " + str((time.time() - start_time)) + ' sec')
```
